```python
import folium
from folium import plugins
import pandas as pd
import numpy as np
from IPython.display import display
```

```python
df = pd.read_csv('df1.csv')
```

```python
map1 = folium.Map()
#display(map1)
"""Important Parameters
    location - where the map shows
    tiles = the style of the map
        - “OpenStreetMap”, “Mapbox Bright”, “Mapbox Control Room”, “Stamen”, “CartoDB Positron”, "CartoDB dark_matter"
    zoom_start = zoom value when the map is made
        - Max zoom is 18, Min zoom is 1
    width 
    height
    control_scale - allows user to manually control zoom once map is made"""
map1 = folium.Map(location=[38, -122.5], tiles='CartoDB positron', zoom_start=8,width=700,height=500,control_scale=True)
map1
```

```python
#Features to add to maps
map1 = folium.Map(location=[38, -122.5], tiles='CartoDB positron', zoom_start=8,width=700,height=500,control_scale=True)
## Displays Lat/Long on the map when clicked
folium.LatLngPopup().add_to(map1)
map1
```

```python
map1 = folium.Map(location=[38, -122.5], tiles='CartoDB positron', zoom_start=8,width=700,height=500,control_scale=True)
## Add Marker where you mouse click
folium.ClickForMarker().add_to(map1)
map1
```

```python
map1 = folium.Map(location=[37.2041, -122.0856],tiles = 'CartoDB positron',zoom_start=8,width=900,height=600,control_scale=True)
#Markers and Lines
## add a marker at the given location
folium.Marker(location=[37.8575,-122.2394 ],icon=folium.Icon(icon='glyphicon-road',prefix='glyphicon',color='orange',icon_color='green'), popup='San Francisco',tooltip='A tooltip!').add_to(map1)
folium.RegularPolygonMarker(location=[37.7447, -121.7889],number_of_sides=10,color='blue',radius=3).add_to(map1)
folium.CircleMarker(location=[38.0567, -122.6898],opacity=1,radius=1,popup='CircleMarker').add_to(map1)        

##icon must be a folium.Icon() object
# adds a lines between 2 points
latlngs = [ [37.8, -122.5],[37.8,-122] ] # A list of lists!
path = folium.PolyLine(latlngs,color='blue',weight=2,popup='A line!',tooltip='First Line').add_to(map1)
display(map1)
```

```python
#Plotting Points from a pandas dataframe
map1 = folium.Map(location=[37.2041, -122.0856],tiles = 'CartoDB positron',zoom_start=8,width=900,height=600,control_scale=True)


for index,row in df.iterrows():
    folium.CircleMarker(location=[row['Latitude'],row['Longitude']],opacity=1,radius=1,popup=row['VehicleId'], tooltip=row['Time']).add_to(map1)        
map1
```

```python
#Plotting lines for each Vehicle
map1 = folium.Map(location=[37.2041, -122.0856],tiles = 'CartoDB positron',zoom_start=8,width=900,height=600,control_scale=True)
for i in df['VehicleId'].unique():
    _data = df[df['VehicleId'] == i]
    for j in range(_data.shape[0]-1):
        LatLng = [[_data.iloc[j]['Latitude'],_data.iloc[j]['Longitude']],
                 [_data.iloc[j+1]['Latitude'],_data.iloc[j+1]['Longitude']]]
        folium.PolyLine(LatLng, color='orange', weight=2, popup=_data.iloc[j]['VehicleId'], tooltip=_data.iloc[j]['Time']).add_to(map1)
map1
```
